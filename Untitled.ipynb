```python
import Alexnet
import torchvision.transforms as transforms
from torch.autograd import Variable
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline



image=cv2.imread('Lotte.jpg')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

alexnet = Alexnet.alexnet(pretrained=True,num_classes=1000)

transform=transforms.ToTensor()

image=cv2.resize(image,(1512,1512))
im_old=image.copy()
plt.imshow(image)
plt.show()

image=transform(image)
image=image.unsqueeze(0)
```

```python
target = torch.tensor((), dtype=torch.int32)
target=target.new_zeros((1, 1000))
target[0][50]=1.00
target=target.float()

def denorm(ima):
    im=ima.clone()
    im=im.detach().squeeze()
    im=im.numpy()*255
    im=np.swapaxes(im,0,2)
    im=np.swapaxes(im,0,1)
    return im.astype(np.uint8)

lr=0.01
image=Variable(image,requires_grad=True)
for epoch in range(5):
    alexnet.zero_grad()
    out=alexnet(image)
    loss=out.norm()
    loss.backward()
    avg_grad = np.abs(image.grad.data.cpu().numpy()).mean()
    norm_lr = lr / avg_grad
    image.data += norm_lr * image.grad.data
    #image.data = clip(image.data)
    image.grad.data.zero_()
im=denorm(image)  
plt.imshow(im)
plt.show()
```
